```python
import docx2txt
import csv
import PyPDF2
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from google.colab import files
```

```python

def extract_text_from_pdf(pdf_file):
    text = ""
    with open(pdf_file, "rb") as file:
        pdf_reader = PyPDF2.PdfReader(file)
        for page in pdf_reader.pages:
            text += page.extract_text()
    return text
```

```python
#Load the data
uploaded_ads = files.upload()
```

```python
# Load the job ads from the CSV file
job_ads = []
with open('ads_data_cleaned.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Skip the first row
    for row in reader:
        job_ads.append(row[2])  # the job description are in the first column
```

```python
# Upload and process the resume
uploadResume = files.upload()


```

```python
resume_file_name = next(iter(uploadResume))
if resume_file_name.endswith(".pdf"):
    resume = extract_text_from_pdf(resume_file_name)
elif resume_file_name.endswith(".docx"):
    resume = docx2txt.process(resume_file_name)
else:
    print("Unsupported file format. Only .pdf and .docx are supported.")
    resume = ""
```

```python
# Prepare the list for storing similarity scores
similarity_scores = []

# Calculate the similarity for each job ad
for job_ad in job_ads:
    # Create a list of resume and current job ad
    text = [resume, job_ad]
    # Perform count vectorization and similarity calculation
    cv = CountVectorizer()
    count_matrix = cv.fit_transform(text)
    similarity_score = cosine_similarity(count_matrix)[0][1]
    similarity_scores.append(similarity_score)

# Sort similarity_scores in descending order and get the indices of the top 10 scores
top_20_indices = sorted(range(len(similarity_scores)), key=lambda i: similarity_scores[i], reverse=True)[:20]

# Print the top 10 similarity scores for each job ad
for i in top_20_indices:
    score = similarity_scores[i]
    match_percentage = round(score * 100, 2)
    print("Job Ad {} matches about {}% of the resume.".format(i + 1, match_percentage))
```
